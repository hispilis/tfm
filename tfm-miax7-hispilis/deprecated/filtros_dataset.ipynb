```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/outputs/dataset_2021_news.csv", sep=";", parse_dates=["date"], index_col=0)
df.shape
```

```python
df['date_agg'] = df.date.dt.date
```

```python
df_grpby = df.groupby(['date_agg','ticker','resource'])
# a = df_paragraph_grpby.apply(lambda x: func_apply(x)).to_frame(name='vola_label_prob')
df_agg = df_grpby.agg({'ticker':'count'})
# df_paragraph_agg = b.join(a)
print(df_agg.shape)
print(df_agg)
df_agg = df_agg[df_agg.ticker==1]
print(df_agg.shape)
df_agg
```

```python
new_df = pd.merge(df, df_agg,  how='inner', left_on=['date_agg','ticker','resource'], right_index=True)
new_df
```

```python
new_df.columns
```

```python
new_df = new_df[new_df.finanzas_prob > 0]
new_df = new_df[new_df.intensidad > 1]
```

```python
from sklearn.utils import shuffle
new_df = shuffle(new_df, random_state=42)
```

```python
new_df.to_csv("../../data/outputs/filters_dataset_2021_news.csv", sep=";")
```
