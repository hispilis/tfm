```python
from subject_classification_spanish import subject_classifier
classifier = subject_classifier.SubjectClassifier()
```

```python
def func_classify(body):    
    return classifier.classify(body)
```

```python
import pandas as pd
df = pd.read_csv("../../data/outputs/output_ner_2021_01_full.csv", sep=";", index_col=0, parse_dates=["date"])
df
```

```python
# df = df[0:100]
from tqdm.auto import tqdm 

tqdm.pandas()
df['topics'] = df['body'].progress_apply(func_classify)
df
```

```python
def is_finanzas(topics):
    FINANZAS_LIST = [ 'economía', 'finanzas', 'bolsa', 'mercados financieros', 'bolsas internacionales']    
    for topic,v in topics.items():
        if topic in FINANZAS_LIST:
            return v
    return 0
```

```python
def best_topic(topics):    
    v_ret = 0.0
    topic_ret = ''
    for k,v in topics.items():
        if v > v_ret:
            v_ret = v
            topic_ret = k
    return topic_ret
```

```python
import numpy as np
def sector_in_topics(row):    
    v_ret = 0.0
    topic_ret = ''    
    if (not pd.isna(row.sector)):
        for k,v in row.topics.items():
            if k in row.sector.lower() or k in row.subsector.lower():
                return 1
    return 0
```

```python
df['sector_in_topics'] = df.apply(lambda row : sector_in_topics(row), axis=1)
df['topic'] = df['topics'].apply(best_topic)
df['finanzas_prob'] = df['topics'].apply(is_finanzas)
df
```
