```python
BLACKLIST = ['ev', 'covid', 'etf', 'nyse', 'sec', 'spac', 'fda', 'treasury']

def get_orgs(text):
    doc = nlp(text)
    org_list = []
    for entity in doc.ents:
        # here we modify the original code to check that entity text is not equal to one of our 'blacklisted' organizations
        # (we also add .lower() to lowercase the text, this allows us to match both 'nyse' and 'NYSE' with just 'nyse')
        if (entity.label_ == 'ORG' or entity.label_ == 'PER') and entity.text.lower() not in BLACKLIST:
            org_list.append(entity.text)
    # if organization is identified more than once it will appear multiple times in list
    # we use set() to remove duplicates then convert back to list
    # org_list = list(set(org_list))
    return org_list
```

```python
# df = df_mini.iloc[0:5000,:]
df = df_mini.copy()

from tqdm.auto import tqdm 

tqdm.pandas()
df['organizations'] = df['body'].progress_apply(get_orgs)
```

```python
orgs = df['organizations'].to_list()
orgs = [org for sublist in orgs for org in sublist]
orgs[:10]
```

```python
import numpy as np

def get_tickers(orgs):    
    ticker_list = []
    for org in orgs:
        #si el ticker esta 
        for k,v in df_tickers['keywords'].to_dict().items():
            l_v = v.split(',')            
            if org in l_v or org in [x.upper() for x in l_v]:                                
                ticker_list.append(k)
                break
        #si el ceo esta
        for k,v in df_ceos['peoples'].to_dict().items():
            if org == v:                
                ticker_list.append(k)
                break            
    # if organization is identified more than once it will appear multiple times in list
    # we use set() to remove duplicates then convert back to list
    # org_list = list(set(org_list))
    ticker_freq = Counter(ticker_list)
    return ticker_freq    
```

```python
from collections import Counter

def get_ticker_info(row):
    if (len(row.tickers)>0):
        row["ticker_freq"] = row.tickers[row.ticker]        
        if len(row.tickers)>1 and list(row.tickers.keys())[0] == row.ticker:
            row["ticker_first"] = 1
        else:
            row["ticker_first"] = 0
        for k,v in df_tickers['keywords'].to_dict().items():                                    
            if (k==row.ticker):
                l_v = v.split(',')
                for el_v in l_v:
                    if el_v in row.title:                        
                        row["ticker_title"] = 1        
                        break
                    else:
                        row["ticker_title"] = 0
                break
            else:
                row["ticker_title"] = 0            
        row["ticker_name"] = df_tickers.loc[row.ticker].legal_name  
        row["sector"] = df_tickers.loc[row.ticker].sector
        row["subsector"] = df_tickers.loc[row.ticker].subsector        
        return row
    else:
        row["ticker_freq"] = 0
        row["ticker_name"] = ''  
        row["sector"] = ''
        row["subsector"] = ''
        return row
```

```python
df["tickers"] = df['organizations'].apply(get_tickers)
df
```
